```python
# scrape feelings, helps_with, and negatives section on leafly.com 

sections = ["feelings", "helps_with", "negatives"]
cannabis_dict = defaultdict(dict)

for strain in strains:
    strain = strain.lower()
    cannabis_dict[strain] = defaultdict(list)
    try:
        index = 0
        page = requests.get(f"https://www.leafly.com/strains/{strain}")
        html_doc = page.text
        soup = BeautifulSoup(html_doc, 'html.parser')
        for section in soup.find_all('section'):
            if 'pt-xxl' in section['class'] and ("Effects" in section.h2):
                for div in section.find_all('div'):
                    if "id" in div.attrs:
                        for d in div.find_all("div"):
                            if "font-bold" in d["class"]:
                                content = d.text
                                content = content.replace("%", "").split()
                                if len(content) > 2:
                                    vals_to_join = len(content) - 1
                                    content = (" ".join(content[:vals_to_join]), int(content[-1]))
                                else:
                                    content = (content[0], int(content[1]))
                                cannabis_dict[strain][sections[index]].append(content)
                        index += 1

        
        cannabis_dict[strain] = dict(cannabis_dict[strain])
        print(f"Info for strain {strain}: {cannabis_dict[strain]}")
        
    except Exception as ex:
        print(ex)
        
pprint(cannabis_dict)
```

```python
result = {
    k: v for k, v in cannabis_dict.items() if v
}
pprint(result)
```
